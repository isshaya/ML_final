Loading Data

```python
import pandas as pd

df_overview = pd.read_csv("all-task-1-overview.csv")
df_reviews = pd.read_csv("all-task-1-featured-reviews.csv")

print("Overview data:")
print(df_overview.columns)
print("%d entries, and %d attributes \n" %(df_overview.shape[0],df_overview.shape[1]))


print("Reviews data:")
print(df_reviews.columns)
print("%d entries, and %d attributes \n" %(df_reviews.shape[0],df_reviews.shape[1]))

# Example Analysis
avg_ratings = df_reviews.groupby("place_id")["rating"].mean()
print("Average ratings per place_id:")
print(avg_ratings)

```

Merging Dataframe

```python
data = pd.merge(df_overview, df_reviews, on="place_id", how="inner")
print("Merged data:")
print(data.info())
print(data.describe(include='all'))
print(data.isnull().sum())
```
